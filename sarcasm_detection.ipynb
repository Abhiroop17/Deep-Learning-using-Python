```python
from simpletransformers.classification import ClassificationModel
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your dataset
df = pd.read_csv('/content/json-to-csv.csv')

# Split the dataset
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Define the RoBERTa model
model = ClassificationModel(
    'roberta',
    'roberta-base',
    num_labels=2,  # Specify the number of classes for your task
    args={'reprocess_input_data': True, 'overwrite_output_dir': True},
    use_cuda= False
)

# Fine-tune the model
model.train_model(train_df)

# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(test_df)
print(result)

# Make predictions
text_to_predict = "This is a sample text for prediction."
predictions, raw_outputs = model.predict([text_to_predict])
print(predictions)
```

```python
# Example prediction
text_to_predict = "This is a sarcastic sentence."
predictions, raw_outputs = model.predict([text_to_predict])

# Display the predictions
print(predictions)
```
